```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# URL страницы Wikipedia
url = 'https://en.wikipedia.org/wiki/2024–25_UEFA_Champions_League_league_phase'

# Отправить GET-запрос к URL
response = requests.get(url)

# Разобрать HTML-контент с помощью BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Найти все таблицы в HTML-контенте
tables = soup.find_all('table')

# Преобразовать таблицы в DataFrame pandas
all_tables = [pd.read_html(str(table))[0] for table in tables]

# Вывести таблицы
for i, table in enumerate(all_tables):
    print(f"Таблица {i+1}:")
    print(table)
    print("\n")
```

```python
dict_table = {}
df = all_tables[7]
df.pop('Qualification')
dict_table = df
```
