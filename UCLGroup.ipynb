```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# URL  Wikipedia
url = 'https://en.wikipedia.org/wiki/202425_UEFA_Champions_League_league_phase'

#  GET-  URL
response = requests.get(url)

#  HTML-   BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

#     HTML-
tables = soup.find_all('table')

#    DataFrame pandas
all_tables = [pd.read_html(str(table))[0] for table in tables]

#  
for i, table in enumerate(all_tables):
    print(f" {i+1}:")
    print(table)
    print("\n")
```

```python
dict_table = {}
df = all_tables[7]
df.pop('Qualification')
dict_table = df
```
